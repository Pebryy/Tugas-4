```python
# 1. Create an array and traverse. ## Membuat array dan menelusuri isinya

from array import array  ## Mengimpor kelas array dari modul array

my_array = array('i', [1,2,3,4,5])  ## Membuat array tipe integer ('i') dengan beberapa elemen
for i in my_array:  ## Loop setiap elemen dalam array
    print(i)  ## Menampilkan setiap elemen
```

```python
# 2. Access individual elements through indexes  ## Mengakses elemen array berdasarkan index

from array import array  ## Mengimpor modul array

print("Step 2")  ## Menampilkan teks penanda langkah
print(my_array[3])  ## Mengakses dan menampilkan elemen pada index ke-3 (elemen ke-4)
```

```python
# 3. Append any value to the array using append() method  ## Menambahkan elemen ke array

from array import array  ## Mengimpor modul array

print("Step 3")  ## Menampilkan teks penanda langkah
my_array.append(6)  ## Menambahkan nilai 6 ke akhir array
print(my_array)  ## Menampilkan array setelah penambahan
```

```python
# 4. Insert value in an array using insert() method  ## Menyisipkan elemen ke dalam array

from array import array  ## Mengimpor modul array

print("Step 4")  ## Menampilkan teks penanda langkah
my_array.insert(3, 11)  ## Menyisipkan nilai 11 pada index ke-3
print(my_array)  ## Menampilkan array setelah penyisipan
```

```python
# 5. Extend python array using extend() method  ## Menambahkan beberapa elemen dari array lain

from array import array  ## Mengimpor modul array

print("Step 5")  ## Menampilkan teks penanda langkah
my_array1 = array('i', [10,11,12])  ## Membuat array baru
my_array.extend(my_array1)  ## Menambahkan semua elemen dari my_array1 ke my_array
print(my_array)  ## Menampilkan array setelah digabung
```

```python
# 13. Convert array to string using tostring() method  ## Mengubah array menjadi string (bytes)

from array import array  ## Mengimpor modul array

print("Step 13")  ## Menampilkan teks penanda langkah
strTemp = my_array.tobytes()  ## Mengubah array menjadi bentuk string/bytes
print(strTemp)  ## Menampilkan hasil string

ints = array('i')  ## Membuat array baru
ints.frombytes(strTemp)  ## Mengisi array dari string/bytes tadi
print(ints)  ## Menampilkan array hasil konversi
```
