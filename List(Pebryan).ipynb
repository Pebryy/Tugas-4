```python
# Update/Insert - List  ## Judul: operasi update dan insert pada list

myList = [1,2,3,4,5,6,7]  ## Membuat list berisi angka 1 sampai 7
print(myList)  ## Menampilkan isi list
```

```python
myList.insert(4, 15)  ## Menyisipkan nilai 15 pada index ke-4 (posisi ke-5)
print(myList)  ## Menampilkan list setelah penambahan elemen
```

```python
myList.append(55)  ## Menambahkan nilai 55 ke akhir list
print(myList)  ## Menampilkan list setelah penambahan elemen
```

```python
newList = [11, 12, 13, 14]  ## Membuat list baru yang akan digabungkan
myList.extend(newList)  ## Menambahkan semua elemen dari newList ke myList

print(myList)  ## Menampilkan list setelah digabung
```

```python
#  Searching for an element in the List  ## Judul: mencari elemen dalam list

myList = [10,20,30,40,50,60,70,80,90]  ## Membuat list berisi beberapa angka

def searchinList(list, value):  ## Fungsi untuk mencari nilai dalam list
    for i in list:  ## Loop setiap elemen dalam list
        if i == value:  ## Mengecek apakah elemen sama dengan nilai yang dicari
            return list.index(value)  ## Mengembalikan index dari nilai tersebut
    return 'The value does not exist in the list'  ## Jika tidak ditemukan

print(myList)  ## Menampilkan isi list
```

```python
print(searchinList(myList, 60))  ## Mencari nilai 60 dalam list dan menampilkan index-nya jika ditemukan
```

```python
#  List operations / functions  ## Judul: menghitung rata-rata dari input user

total = 0  ## Menyimpan total penjumlahan
count = 0  ## Menyimpan jumlah data

while (True):  ## Loop terus sampai kondisi berhenti
    inp = input('Enter a number: ')  ## Meminta input dari user
    if inp == 'done': break  ## Jika user mengetik 'done', loop berhenti
    value = float(inp)  ## Mengubah input menjadi tipe float (angka desimal)
    total = total + value  ## Menambahkan nilai ke total
    count = count + 1  ## Menambah jumlah data
    average = total / count  ## Menghitung rata-rata sementara

print('Average:', average)  ## Menampilkan hasil rata-rata
```

```python
numlist = list()  ## Membuat list kosong untuk menyimpan angka

while (True):  ## Loop terus sampai user berhenti
    inp = input('Enter a number: ')  ## Meminta input dari user
    if inp == 'done': break  ## Menghentikan loop jika user mengetik 'done'
    value = float(inp)  ## Mengubah input menjadi angka (float)
    numlist.append(value)  ## Menambahkan nilai ke dalam list

average = sum(numlist) / len(numlist)  ## Menghitung rata-rata dari semua elemen dalam list

print('Average:', average)  ## Menampilkan hasil rata-rata
```
