```python
# Q-3. What will be the output of the following code snippet?

a = [1,2,3,4,5]  ## Membuat list

print(a[3:0:-1])  ## Mulai dari index 3 (nilai 4), bergerak mundur (step -1) sampai sebelum index 0
                  ## Elemen yang diambil: index 3 → 2 → 1 (tidak termasuk index 0)

##JAWABAN : [4, 3, 2]
```

```python
# Q-4. What will be the output of the following code snippet?

def f(value, values):
    v = 1  ## Variabel lokal (tidak mempengaruhi luar)
    values[0] = 44  ## Mengubah isi list (mutable → berdampak ke luar)

t = 3  ## Tipe integer (immutable)
v = [1, 2, 3]  ## List (mutable)

f(t, v)  ## Memanggil fungsi

print(t, v[0])  ## t tidak berubah, v[0] berubah jadi 44

##JAWABAN : C

# A. 1 44
# B. 3 1
# C. 3 44
# D. 1 1
```

```python
# Q-6. What will be the output of the following code snippet?

data = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]

def fun(m):
    v = m[0][0]  ## Ambil nilai awal (1)

    for row in m:  ## Loop setiap sub-list
        for element in row:  ## Loop setiap elemen
            if v < element: v = element  ## Update jika menemukan nilai lebih besar

    return v  ## Mengembalikan nilai terbesar

print(fun(data[0]))  ## Mengirim [[1,2],[3,4]]

##JAWABAN : D

# A. 1
# B. 2
# C. 3
# D. 4
# E. 5
# F. 6
```

```python
# Q-8. What will be the output of the following code snippet?

def f(i, values = []):
    values.append(i)  ## Menambahkan nilai ke list yang sama (tidak reset)
    print(values)
    return values

f(1)
f(2)
f(3)

##JAWABAN : C

# A. [1] [2] [3]
# B. [1, 2, 3]
# C. [1] [1, 2] [1, 2, 3]
# D. 1 2 3
```

```python
# Q-9. What will be the output of the following code snippet?

arr = [1, 2, 3, 4, 5, 6]

for i in range(1, 6):
    arr[i - 1] = arr[i]  ## Menggeser elemen ke kiri (menimpa dengan elemen berikutnya)
for i in range(0, 6):
    print(arr[i], end = " ")  ## Menampilkan hasil akhir

##JAWABAN : D

# A. 1 2 3 4 5 6
# B. 2 3 4 5 6 1
# C. 1 1 2 3 4 5
# D. 2 3 4 5 6 6
```

```python
# Q-10. What will be the output of the following code snippet?

fruit_list1 = ['Apple', 'Berry', 'Cherry', 'Papaya']
fruit_list2 = fruit_list1  ## Reference (terhubung ke list yang sama)
fruit_list3 = fruit_list1[:]  ## Copy (list baru)

fruit_list2[0] = 'Guava'  ## Mengubah fruit_list1 juga
fruit_list3[1] = 'Kiwi'  ## Hanya mengubah fruit_list3

sum = 0

for ls in (fruit_list1, fruit_list2, fruit_list3):  ## Loop semua list
    if ls[0] == 'Guava':  ## Cek elemen pertama
        sum += 1
    if ls[1] == 'Kiwi': ## Cek elemen kedua
        sum += 20

print(sum)

##JAWABAN : A

# A. 22
# B. 21
# C. 0
# D. 43
```
